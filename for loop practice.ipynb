```python
# print the multification table of a number(take input from user)

n=int(input("enter your number:-"))

for i in range(1,11):
    print(f"{n} x {i} = {n*i}")
```

```python
# calculate the sum of the number form 1 to n(input from user)

n=int(input("enter your number:-"))

sum=0

for i in range(1,n+1):
      sum=sum+i
print(sum)
```

```python
# count the number of vowels in a given string.
name=input("Enter your name:-")
vowels="aeiouAEIOU"
count=0

for char in name:
  if char in vowels:
    count+=1
print(f"number of the count {count}")
```

```python
# Find the factorial of the number
n=int(input("Enter your number:-"))
fact=1

for i in range(1,n+1):
    fact *=i

print(fact)
```
